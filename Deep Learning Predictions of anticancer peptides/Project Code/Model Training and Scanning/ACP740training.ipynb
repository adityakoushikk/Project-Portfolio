```python
#import numpy as np
from sklearn.metrics import matthews_corrcoef
def Recall(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def Precision(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def F1(y_true, y_pred):
    true_negatives = K.sum(K.round(K.clip((1 - y_true) * (1 - y_pred), 0, 1)))
    false_positives = K.sum(K.round(K.clip((1 - y_true) * y_pred, 0, 1)))
    specificity = true_negatives / (true_negatives + false_positives + K.epsilon())
    return specificity


def MCC(y_true, y_pred):
    y_pred_pos = K.round(K.clip(y_pred, 0, 1))
    y_pred_neg = 1 - y_pred_pos

    y_pos = K.round(K.clip(y_true, 0, 1))
    y_neg = 1 - y_pos

    tp = K.sum(y_pos * y_pred_pos)
    tn = K.sum(y_neg * y_pred_neg)

    fp = K.sum(y_neg * y_pred_pos)
    fn = K.sum(y_pos * y_pred_neg)

    numerator = (tp * tn - fp * fn)
    denominator = K.sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn))

    return numerator / (denominator + K.epsilon())
```

```python
from sklearn.model_selection import StratifiedKFold


from tensorflow.keras import models, layers, utils, backend as K
kfold = StratifiedKFold(n_splits=5, shuffle=True)
accuracy = []
mcc = []
f1 = []
recall = []
precision= []
all_peptide_pred = []
n_features = 478

for train, test in kfold.split(X_normalized, y):
    
    model = models.Sequential(name="DeepNN", layers=[
        ### hidden layer 1
        layers.Dense(name="h1", input_dim=n_features,
                     units=int(round((n_features+1)/2)), 
                     activation='relu'),
        layers.Dropout(name="drop1", rate=0.6),
      

        ### hidden layer 2
        layers.Dense(name="h2", units=int(round((n_features+1)/4)), 
                     activation='relu'),
        layers.Dropout(name="drop2", rate=0.5),


        layers.Dense(name="h3", units=int(round((n_features+1)/6)), 
                     activation='relu'),
        layers.Dropout(name="drop3", rate=0.2),
 



        ### layer output
        layers.Dense(name="output", units=1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy',MCC,F1,Recall,Precision ])
    # Fit the model
    model.fit(X_normalized[train], y[train], epochs=20, batch_size=90, verbose=0)
    scores = model.evaluate(X_normalized[test], y[test], verbose=1)
    accuracy.append(scores[1] * 100)
    mcc.append(scores[2] * 100)
    f1.append(scores[3] * 100)
    recall.append(scores[4] * 100)
    precision.append(scores[5] * 100)

   


    




```

```python
import statistics
def StandardDeviation(lst):
    return statistics.stdev(lst)
# default
def Average(lst):
    return sum(lst) / len(lst)
```
